```python
import pandas as pd

# Create the DataFrame with subject marks
data = {
    'Subject_name': ['Mathematics', 'Physics', 'Chemistry', 'English', 'Computer Science'],
    'Subdivision': ['Science', 'Science', 'Science', 'Arts', 'Science'],
    'Year': [2019, 2020, 2020, 2021, 2021],
    'Semester': [1, 2, 1, 2, 1],
    'Marks_scored': [80, 75, 90, 85, 95],
    'Max_marks': [100, 100, 100, 100, 100],
    'Grade': ['A', 'B', 'A', 'B', 'A']
}

df = pd.DataFrame(data)

# Define the user-defined function to categorize marks
def categorize_marks(marks):
    if marks > 90:
        return 'Excellent'
    elif marks > 80:
        return 'Very Good'
    elif marks > 70:
        return 'Good'
    else:
        return 'Average'

# Apply the function to the 'Marks_scored' column and create a new column 'Mark_Category'
df['Mark_Category'] = df['Marks_scored'].apply(categorize_marks)

# Remove the 'Max_marks' column
df = df.drop('Max_marks', axis=1)

# Move the 'Mark_Category' column after the 'Marks_scored' column
columns = df.columns.tolist()
columns.insert(columns.index('Marks_scored') + 1, 'Mark_Category')
df = df[columns]

print(df)
```

```python
df['Marks_category']=df['Marks_scored'].apply(lambda x: 'Excellent' if x>90 else 'Very_good' if x>=80 else 'Good' if x>=70 else 'Average')
print(df)
```

```python
sem_stats=df.groupby('Semester')['Marks_scored'].agg('mean','median','mode')
sem_stats
```

```python
sem_avg=df.groupby('Semester')['Marks_scored'].mean()
```

```python
pyplot.plot(sem_avg.index,sem_avg.values,marker='*')
```

```python
pyplot.xlabel('Sem_avg')
```

```python
import matplotlib.pyplot as pyplot

# Calculate semester-wise average marks
semester_avg = df.groupby('Semester')['Marks_scored'].mean()

# Create a line plot
pyplot.plot(semester_avg.index, semester_avg.values, marker='*')
pyplot.xlabel('Sem')
pyplot.ylabel('avg_marks')
pyplot.title('Avg_sem_marks')
pyplot.grid(True)
pyplot.show()
```
